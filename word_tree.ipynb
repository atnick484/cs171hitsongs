```python
df_billboard = pd.read_csv('data/df_billboard_0_10.csv')
```

```python
text_list = df_billboard['lyrics_text'].to_list()
text_list = [text if text is not np.nan else '' for text in text_list]
```

```python
def find_ngrams(text, n):
    if type(text) == float:
        print(text)
    tokens = text.split()
    if n == 1:
        return tokens
    return list(zip(*[tokens[i:] for i in range(n)]))

def create_ngrams_dict(text, n_range):
    ngrams_dict = dict()
    for n in n_range:
        ngrams_dict[n] = find_ngrams(text, n)
    return ngrams_dict
```

```python
all_ngram_dicts = []
for text in text_list:
    all_ngram_dicts.append(create_ngrams_dict(text, range(1, 6)))
df_billboard['ngrams'] = all_ngram_dicts
```

```python
def get_ngram_counts():
    ngrams = df_billboard['ngrams'].to_list()
    combined_ngrams = dict(zip(range(1, 6), [[],[],[],[],[]]))
    for ngram in ngrams:
        keys = ngram.keys()
        for key in keys:
            combined_ngrams[key] += ngram[key]
    for key, value in combined_ngrams.items():
        combined_ngrams[key] = Counter(value).most_common()
    return combined_ngrams
```
