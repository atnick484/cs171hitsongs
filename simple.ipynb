```python
from lyricsgenius import Genius

genius = Genius(
    'in424DYZRH1oa--Iyzce_Ue6uLYNqt1Q6-Yq20PUSYl-PjvMNuuaX0YBiwcSFt9_',
    retries=50,
    verbose=False
)
```

```python
song_parts_regex = r"\[Intro.*?\]|\[Verse [0-9].*?\]|\[Refrain.*?\]|\[Pre-Chorus.*?\]|\[Chorus.*?\]|\[Post-Chorus.*?\]|\[Hooks.*?\]|\[Riffs/Basslines.*?\]|\[Scratches.*?\]|\[Bridge.*?\]|\[Interlude.*?\]|\[Break.*?\]|\[Skit.*?\]|\[Collision.*?\]|\[Instrumental or Solo.*?\]|\[Ad lib.*?\]|\[Segue.*?\]|\[Outro.*?\]"
```

```python
def get_chorus(song, artist):
    raw_lyrics_list = genius.search_song(song, artist).lyrics.split('\n')
    filtered_lyrics_list = filter(lambda line: line != '', raw_lyrics_list)
    chorus = []
    chorus_indicator = False
    for line in filtered_lyrics_list:
        if chorus_indicator:
            if re.match(song_parts_regex, line):
                break
            chorus.append(line)
        if re.match(r"\[Chorus.*?\]", line):
            chorus_indicator = True
    return chorus
```

```python
song_artist = pd.read_csv('test_data/song_artist_gold.csv')
song_artist = [(song, artist) for song, artist in song_artist.to_numpy()]
song_artist[:5]
```

```python
def get_genius_song_object(song, artist):
    song_object = genius.search_song(song, artist)
    if song_object:
        return song_object
    else:
        print(song, artist)
        return None
```

```python
def write_song_chunk_to_json(start, stop):
    song_artist_chunk = song_artist[start:stop]
    song_dicts = []
    for song, artist in song_artist_chunk:
        genius_song_object = get_genius_song_object(song, artist)
        if genius_song_object:
            song_dicts.append(genius_song_object.to_dict())
    with open(f'test_data/song_jsons_{start}_to_{stop}.json', 'x') as json_file:
        json.dump(song_dicts, json_file)
```

```python
def write_song_chunk_to_json_starting_from(start=0, stop=len(song_artist), skip=100):
    ranges = list(zip(list(range(start, stop, skip)), list(range(start + skip, stop + skip, skip))))
    for this_start, this_stop in ranges:
        write_song_chunk_to_json(this_start, this_stop)
```
